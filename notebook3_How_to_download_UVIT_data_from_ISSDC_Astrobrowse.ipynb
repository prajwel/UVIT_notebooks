The [ISSDC Astrobrowse website](https://astrobrowse.issdc.gov.in/astro_archive/archive/Home.jsp) is presently the only archive where AstroSat UVIT data can be accessed publicly. You need to have a username and password to log in to the Astrobrowse website. Click on the register link to create login credentials if you do not already possess one. Once you have logged in to the website, search and download your source of interest. For example, we will search for the data of NGC 7252 (also known as Atoms for Peace Galaxy). Please see the GIF movie below. We will find an Observation ID `G07_003T02_9000001564` (PI: Koshy) matching our search criteria. This data is now publicly available for download after the proprietary period. Two publications are associated with this data. See,
* George, K., et al. "Dissecting star formation in the "Atoms-for-Peace" galaxy-UVIT observations of the post-merger galaxy NGC7252." Astronomy & Astrophysics 614 (2018): A130. 
* George, K., et al. "UVIT observations of the star-forming ring in NGC 7252: Evidence of possible AGN feedback suppressing central star formation." Astronomy & Astrophysics 613 (2018): L9.

![](images/AstroBrowse-fast.gif)

The data download may take a while to complete (I get ~0.5 MB/s bandwidth from ISSDC, so a 2 GB dataset takes ~1 hour).

After downloading the Level2 data (Level1: unprocessed, Level2: processed), we will have a zipped directory, `LEVL2AS1UVT20170927G07_003T02_9000001564.zip`. Once extracted, a directory named `20170927_G07_003T02_9000001564_level2` can be found. This directory has the following structure.
```
20170927_G07_003T02_9000001564_level2/
└── uvit
    ├── DISCLAIMER.txt
    ├── LEVL1AS1UVT20170927G07_003T02_9000001564_10811_V2.2_dqr.xml
    ├── RAS_NUV
    │   ├── pipeline
    │   ├── uvt_01
    │   ├── uvt_02
    │   ├── uvt_03
    │   ├── uvt_04
    │   ├── uvt_05
    │   ├── uvt_06
    │   └── uvt_ci
    ├── RAS_VIS
    │   ├── pipeline
    │   ├── uvt_01
    │   ├── uvt_02
    │   ├── uvt_03
    │   ├── uvt_04
    │   ├── uvt_05
    │   ├── uvt_06
    │   └── uvt_ci
    └── README.txt

```

![](images/Level2_directories-fast.gif)

Please read the `README.txt` for details on Level2 directory structure and the files it contains. 

* `RAS_VIS` directory contains images corrected for satellite drift using the VIS (visible) channel. 
* `RAS_NUV` directory contains images corrected for satellite drift using the NUV (near-ultraviolet) channel. 

For most cases, the data from `RAS_VIS` would be suitable. Check the statistics inside `DISCLAIMER.txt` to decide what to use. 
 
For this case, data statistics inside the `DISCLAIMER.txt` is as follows. Note that the values are in seconds.  

```
        |	VIS 	|	NUV 	|	FUV
- - —  - - - —  - -  —  - - —  - - - —  - -  — 
ASV     |	8300	|	8300	|	8300
- - —  - - - —  - -  —  - - —  - - - —  - -  — 
L1      |	8770	|	8268	|	8247
- - —  - - - —  - -  —  - - —  - - - —  - -  — 
L2(NUV) |		    |	3180	|	3183
- - —  - - - —  - -  —  - - —  - - - —  - -  — 
L2(VIS) |	        |	8267	|	8247
```

* ASV: This row lists the total duration of scheduled observations in each of the channels of UVIT. See [Astrosat Schedule Viewer (ASV)](https://webapps.issdc.gov.in/MCAP/) website.
* L1: Lists the total unprocessed data found inside Level1 dataset.
* L2 (NUV): Lists the total processed data found inside RAS_NUV orbits/episodes.
* L2 (VIS): Lists the total processed data found inside RAS_VIS orbits/episodes.

We can see that the `RAS_VIS` directory contains more data than `RAS_NUV` from statistics provided in `DISCLAIMER.txt`. 

Our directory of interest, `RAS_VIS`, has the following contents.

```
RAS_VIS/
├── pipeline
│   └── LEVL2AS1UVT20170927G07_003_L2_DM_params.txt
├── uvt_01
│   ├── F_01
│   ├── N_01
│   └── V_01
├── uvt_02
│   ├── F_02
│   ├── N_02
│   └── V_02
├── uvt_03
│   ├── F_03
│   ├── N_03
│   └── V_03
├── uvt_04
│   ├── F_04
│   ├── N_04
│   └── V_04
├── uvt_05
│   ├── F_05
│   ├── N_05
│   └── V_05
├── uvt_06
│   ├── F_06
│   ├── N_06
│   └── V_06
└── uvt_ci
    ├── F_ci
    └── N_ci

```

You will notice that there are directories named in a pattern, `uvt_01`, `uvt_02`, . . . , `uvt_06`. They correspond to different orbits/episodes of the UVIT observation. 

For example, `RAS_VIS/uvt_01` has the following contents.

```
uvt_01
├── F_01
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2err.fits
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2exp.fits
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2img.fits
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1I_l2img.fits
│   └── AS1G07_003T02_9000001564uvtFIIPC00F1_l2ce.fits
├── N_01
│   ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2err.fits
│   ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2exp.fits
│   ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2img.fits
│   ├── AS1G07_003T02_9000001564uvtNIIPC00F1I_l2img.fits
│   └── AS1G07_003T02_9000001564uvtNIIPC00F1_l2ce.fits
└── V_01
    └── AS1G07_003T02_9000001564uvtVIIIM00F2_l2dr.fits
```

Inside the directory `uvt_01`, data are organized in separate folders, each corresponding to overlapping time ranges in UV and VIS channels, as available in Level1 dataset (`F_01`: FUV; `N_01`: NUV; `V_01`: VIS).

The suffixes of the FITS files have the following meaning.

`...A_l2img.fits`: Image file in astronomical coordinates.

`...I_l2img.fits`: Image file in instrument coordinates.

`...A_l2exp.fits`: Exposure map for `A_l2img.fits`.

`...A_l2err.fits`: Error map for `A_l2img.fits`.

`...l2ce.fits` : Corrected events list.

`...l2dr.fits` : the Relative Aspects Series (RAS) file.

This structure of subdirectories shall repeat for all sets - uvt_01, uvt_02, uvt_03, etc.

The directory `uvt_ci` contains the combined images. For example, `RAS_VIS/uvt_ci` has the following contents.

```
uvt_ci
├── F_ci
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2erb.fits
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2exb.fits
│   ├── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2ima.fits
│   └── AS1G07_003T02_9000001564uvtFIIPC00F1A_l2imb.fits
└── N_ci
    ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2erb.fits
    ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2exb.fits
    ├── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2ima.fits
    └── AS1G07_003T02_9000001564uvtNIIPC00F1A_l2imb.fits
```

The suffixes of the FITS files have the following meaning.


`...A_l2imb.fits`: Image file in astronomical coordinates (WCS generated by catalogue matching).

`...A_l2ima.fits`: Image file in astronomical coordinates (WCS generated from spacecraft aspect).

`...A_l2exb.fits`: Exposure map for `A_l2imb.fits`.

`...A_l2erb.fits`: Error map for `A_l2imb.fits`.
